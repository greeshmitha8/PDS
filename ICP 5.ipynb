# Slicing

- .loc performances label-based extraction.
- .iloc performs integer-based extraction.

In pandas, .loc and .iloc are two methods used to access and manipulate data in a DataFrame or Series by selecting specific rows and columns. They are powerful tools for selecting data based on labels (.loc) or index positions (.iloc).

.loc: Label-Based Indexing
.loc is used for selecting data by label (i.e., row and column names).
It allows you to select rows and columns based on their index labels (row/column names).
It can handle both single labels and lists or slices of labels.

.iloc: Position-Based Indexing
.iloc is used for selecting data by integer positions (i.e., row and column indices).
It allows you to access data based on the position of rows and columns rather than their labels.

```python
df.loc[0, 'Loan_ID']
```

```python
df.loc[[68, 100, 201], 'Married']
```

```python
df.iloc[0,4]
```

# Using the Sacramento Housing  data

```python
import pandas as pd

# Load the dataset from the given link
url = 'https://raw.githubusercontent.com/JuliaData/CSV.jl/refs/heads/main/test/testfiles/Sacramentorealestatetransactions.csv'
df = pd.read_csv(url)

# Display the first few rows of the dataset
df.head()
```

## Find the number of the columns in the data frame

```python
num_columns = len(df.columns)
print(f'Number of columns: {num_columns}')
```

## Find the mean of the house prices

```python
mean_price = df['price'].mean()
print(f'Mean of house prices: {mean_price}')
```

## Find the median of the house prices

```python
median_price = df['price'].median()
print(f'Median of house prices: {median_price}')
```

## Find the different types of property in the dataset

```python
property_types = df['type'].unique()
print(f'Different property types: {property_types}')
```

## Find the maximum price of a "condo" in the dataset

```python
max_condo_price = df[df['type'] == 'Condo']['price'].max()
print(f'Maximum price of a condo: {max_condo_price}')
```

## Find the mean of each type of property in the dataset

```python
mean_by_property_type = df.groupby('type')['price'].mean()
print(f'Mean price for each property type:\n{mean_by_property_type}')
```

## Find the count of different property types in dataset

```python
property_type_counts = df['type'].value_counts()
print(f'Count of different property types:\n{property_type_counts}')
```

## Cleaning the dataset and performing same activities

```python
# Check for missing values
print("Missing values before cleaning:\n", df.isnull().sum())

# Drop rows with missing values
df_cleaned = df.dropna()

# Check for duplicates
duplicates = df_cleaned.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')

# Drop duplicates if any
df_cleaned = df_cleaned.drop_duplicates()

# Verify data types
print("\nData types before conversion:\n", df_cleaned.dtypes)

# Convert price to numeric if not already (ensure it's a valid type)
df_cleaned['price'] = pd.to_numeric(df_cleaned['price'], errors='coerce')

# Check for any remaining missing values after conversion
df_cleaned = df_cleaned.dropna(subset=['price'])

# Remove outliers in the price column (if necessary)
# For example, assuming any price over 1 million is considered an outlier
df_cleaned = df_cleaned[df_cleaned['price'] < 1000000]

# Rename columns (optional, based on need)
df_cleaned.columns = df_cleaned.columns.str.strip().str.lower().str.replace(' ', '_')

# Final clean dataset summary
print("\nSummary after cleaning:")
print(df_cleaned.info())
```

```python
# Number of columns in the cleaned DataFrame
num_columns = len(df_cleaned.columns)
print(f'Number of columns: {num_columns}')

# Mean of house prices
mean_price = df_cleaned['price'].mean()
print(f'Mean of house prices: {mean_price}')

# Median of house prices
median_price = df_cleaned['price'].median()
print(f'Median of house prices: {median_price}')

# Different property types in the dataset
property_types = df_cleaned['type'].unique()
print(f'Different property types: {property_types}')

# Maximum price of a "condo"
max_condo_price = df_cleaned[df_cleaned['type'] == 'Condo']['price'].max()
print(f'Maximum price of a condo: {max_condo_price}')

# Mean price for each property type
mean_by_property_type = df_cleaned.groupby('type')['price'].mean()
print(f'Mean price for each property type:\n{mean_by_property_type}')

# Count of different property types
property_type_counts = df_cleaned['type'].value_counts()
print(f'Count of different property types:\n{property_type_counts}')
```
