```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data
data = pd.read_csv("..//cars_clean_data//cars_clean_data.csv")
```

```python
data.head()
```

#### (E) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in        python) on this dataset

```python
data.head()
```

```python
# Filtering data based on a condition

filtered_data = data[data['Location'] == 'Chennai']

filtered_data.head()
```

```python
# Renaming columns
renamed_data = data.rename(columns={'Kilometers_Driven': 'Kms_Driven', 'Price': 'Price_INR'})
```

```python
renamed_data.head()
```

```python
# Sorting the data by a column Year
sorted_data = data.sort_values(by='Year')
```

```python
sorted_data.head()
```

```python
# Summarizing with groupby
summary_data = data.groupby('Fuel_Type').agg({'Price': 'mean'})
```

```python
summary_data.head()
```

### Data Visuvalization

```python
sns.kdeplot(data['Price'], shade=True, color='green')
plt.xlabel('Price (in lakhs)')
plt.title('Density Plot of Car Prices')
plt.show()
```

```python

plt.figure(figsize=(9, 6))
sns.barplot(x='Fuel_Type', y='Price', data=data, estimator=lambda x: sum(x) / len(x))
plt.xlabel('Type of fuel')
plt.ylabel('Average Price')
plt.title('Average Price of Cars by Fuel')
plt.show()
```

```python
import seaborn as sns
sns.violinplot(x=data['Price'], color='skyblue')
plt.xlabel('Price (in lakhs)')
plt.title('Violin Plot of Car Prices')
plt.show()
```
