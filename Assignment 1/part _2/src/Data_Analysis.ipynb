```python
# importing libaries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading data
Students_data = pd.read_csv("..//clean_data//StudentsPerformance_clean.csv")
```

```python
# dispalying top 5 rows of data
Students_data.head()
```

### Data Analysis

#### 1. Boxplot of Reading Scores by Race/Ethnicity
- This visualization illustrates the distribution of reading scores across various demographic groups, highlighting differences and trends that may reveal inequities in educational outcomes. By examining these disparities, we can gain valuable insights into the factors affecting performance among different races and ethnicities.



```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Race_Ethnicity', y='Reading Score', data=Students_data, palette='Set2')
plt.title('Boxplot by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Reading Score')
plt.savefig('..//results//boxplot_score.png')
plt.show()
```

#### 2. Histogram of Math Scores
- The histogram displays the distribution of math scores among students, highlighting central tendency and variability in performance.

```python
plt.figure(figsize=(8, 5))
plt.hist(Students_data['Math Score'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Histogram of Math Score')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.savefig('..//results//histogram_score.png')
plt.show()
```

#### 3. Stacked Bar Chart of Lunch Type by Gender
- The chart shows that more females than males had lunch, with a higher overall count of female students. Standard lunch was the most popular option, while a greater percentage of females opted for free/reduced lunch compared to males. Conversely, a larger proportion of males chose standard lunch over females.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', hue='Lunch', data=Students_data, palette='Set3')
plt.title('Stacked Bar by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Lunch Type')
plt.savefig('..//results//stacked_bar.png')
plt.show()
```

#### 4. Scatter Plot of Reading Scores vs. Writing Scores
- The scatter plot allows for an easy evaluation of the correlation between reading and writing scores, highlighting the consistency of performance across these two skills.

```python
plt.figure(figsize=(8, 5))
plt.scatter(Students_data['Reading Score'], Students_data['Writing Score'], color='red')
plt.title('Scatter Plot_Reading Scores vs.Scatter Plot_Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.savefig('..//results//scatter_plot.png')
plt.show()
```

#### 5. Bar Chart of Test Preparation Course Completion
- This visualization facilitates a straightforward comparison of student performance improvements from test preparation materials, aiding in the assessment of their effectiveness.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Test Preparation Course', data=Students_data, palette='pastel')
plt.title('Bar Chart of Course Completion')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.savefig('..//results//bar_chart.png')
plt.show()
```

```python
# Computing the correlation matrix
corr = Students_data.corr()

colors = ['#9b59b6', '#CC80AC', '#EE82EE']  
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap=colors, fmt=".2f", annot_kws={"size": 12})
plt.title('Correlation Heatmap')
plt.savefig('..//results//heatmap_correlation_.png')  # Saving the figure
plt.show()
```
